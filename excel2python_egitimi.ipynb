```python
#bir sayının asal olup olmadığını anlayan fonksiyon
def asal_mi(n):
    if n<2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i ==0:
            return False
    return True

print(asal_mi(7))
print(asal_mi(10))
print(asal_mi(101))
```

#### CSV Çalışması

```python
path =r'T:\python-anlatim\veri-birlestir'  # Dosyaların bulunduğu klasörü buraya yazalım..
filenames = glob.glob(path + "/*.csv")


dfs1 = []
for filename1 in filenames:
    dfs1.append(pd.read_csv(filename1,sep=" "))  ## bu veride csv TAB ile ayrildigindan "sep" farkli idi..

# Concatenate all data into one DataFrame
big_frame1 = pd.concat(dfs1, ignore_index=True)
big_frame1.head(5)
```

#### Excel Çalışması

```python
path =r'T:\python-anlatim\veri-birlestir'  
filenames = glob.glob(path + "/*.xlsx")


dfs1 = []
for filename1 in filenames:
    dfs1.append(pd.read_excel(filename1,sheet_name="Sheet1"))  ## excelde hangi sayfa birleştirilecekse sheet_name olarak yazılmalı

# Concatenate all data into one DataFrame
big_frame1 = pd.concat(dfs1, ignore_index=True)
big_frame1.head(5)
```
